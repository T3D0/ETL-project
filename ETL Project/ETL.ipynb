```python
# Import dependencies
import pandas as pd
import scapy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# import password so y'all can't see it
from config import mypass
```

```python
csv_file = "museums.csv"
museum_file = pd.read_csv(csv_file, encoding = 'UTF-8')
museum_file.head()
```

```python
new_museum_df = museum_file[['Museum ID', 'Museum Name', 'State (Administrative Location)']]
new_museum_rename = new_museum_df.rename(columns={'Museum ID': 'Museum_ID', 'Museum Name': 'Museum_Name', 'State (Administrative Location)': 'State'})
new_museum_rename.head()
```

```python
new_museum_rename.reset_index(level=0, inplace=True)
new_museum_rename.head()
```

```python
#new_museum_rename.reset_index(level=0, inplace=True)
#new_museum_rename.head()
Museums = new_museum_rename[['Museum_ID', 'Museum_Name', 'State']]
Museums.drop_duplicates('Museum_Name', inplace=True)
Museums.set_index('Museum_Name', inplace=True)
Museums
```

```python
json_file = "traveler_rating_USonly.json"
traveler_rating_df = pd.read_json(json_file)
traveler_rating_df.head()
```

```python
traveler_rating_average = traveler_rating_df.mean(axis=0)
traveler_rating_average.head()
```

```python
print (traveler_rating_average)
```

```python
indexed_traveler = traveler_rating_average.reset_index(level=0)
indexed_traveler.head()
```

```python
renamed_traveler = indexed_traveler.rename(columns={'index': 'Museum_Name', 0: 'Rating'})
renamed_traveler
renamed_traveler.set_index('Museum_Name', inplace=True)
renamed_traveler
```

```python
connection_string = f"root:{mypass}@localhost/Museums_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
Museums.to_sql(name = 'Museum_Directory', con=engine, if_exists='append', index=True)
```

```python
renamed_traveler.to_sql(name = 'Museum_Reviews', con=engine, if_exists='append', index=True)
```
